#### Manuals and reading material:

* [The Python reference manual](https://docs.python.org/3/library/index.html) - the official online Python manual describing all the Python code that is distributed with Python.
* [Think Python](http://greenteapress.com/thinkpython/html/index.html): How to think like a computer scientist - If you like learning from books then this is a good one.
* [Python conferences](https://wiki.python.org/moin/PythonConferences) - a list of  worldwide Python conference you should attend or at least watch the free recordings online. While PyCon is the most popular conference the link gives a complete list of other conferences as well.

#### Community forums:

* [r/python](https://www.reddit.com/r/Python/) - a very active Python community on Reddit that is about Python news, and Python code examples.
* [r/learnpython](https://www.reddit.com/r/learnpython/) - the other very active Python community on Reddit about asking or posting anything that has to do with learning Python.
* [StackOverflow](https://stackoverflow.com/) - A more technical Question & Answer forum that you should use whenever a coding problem arises. Just ask a question there and most of the time you will get an answer in minutes.

#### Online tools:

* [Jupyter Binder](https://mybinder.org/) - Lets you create, run and share Python code online with anyone. You need an account on GitHub for that to work.
* [PythonAnywhere](https://www.pythonanywhere.com/) - Gives you access to a Linux virtual server that lets you write and run Python code. In addition you can schedule your Python programs for execution at certain time every day and you can also deploy your Python web apps on PythonAnywhere.

#### Practice material:

* [List of Python challenges](https://pythonhow.com/link-python-challenges/) - Challenges are a great a way to learn Python. The link contains is a list of five good Python challenges.
* [Python projects](https://github.com/tuvtran/project-based-learning/blob/master/README.md#python) - This is a good list of Python projects in a wide range of application areas.
